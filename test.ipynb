```python
from visualization import Visualizer
from value import Value

visualizer = Visualizer()

x1 = Value(2, 'x1')
x2 = Value(0, 'x2')

w1 = Value(-3, 'w1')
w2 = Value(2, 'w2')

b = Value(6.8813735870159432, 'b')

n = w1*x1 + w2*x2 + b
o = n.tanh(); o.symbol = 'o'

visualizer.draw_diagram(o)
```

```python
o.back_propogate()
visualizer.draw_diagram(o)
```

```python

def build_topological_graph(tree, graph=[], visited=set()):
    if tree not in visited:
        visited.add(tree)
        for child in tree.children:
            build_topological_graph(child, graph, visited)
        graph.append(tree)
    return graph
```

```python
build_topological_graph(o)
```
